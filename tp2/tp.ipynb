```python
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
original_image = cv.imread('washers.jpg')
plt.imshow(original_image)
plt.show()
```

```python
rgb_image = cv.cvtColor(original_image, cv.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.show()
```

```python
hsv_image = cv.cvtColor(rgb_image, cv.COLOR_BGR2HSV)
plt.imshow(hsv_image)
plt.show()
```

```python
colors = {
    "red": {
        "lower_bound": (119,100,100),
        "upper_bound": (125,255,255),
    },
    "blue": {
        "lower_bound": (2,100,100),
        "upper_bound": (8,255,255),
    },
    "orange": {
        "lower_bound": (106,100,100),
        "upper_bound": (108,255,255),
    },
    "light_blue": {
        "lower_bound": (12,100,100),
        "upper_bound": (15,255,255),
    }
}
```

```python
for key in colors:
    mask = cv.inRange(hsv_image, colors[key]["lower_bound"], colors[key]["upper_bound"])
    masked_image = cv.bitwise_and(rgb_image, rgb_image, mask=mask)

    print(key)
    plt.figure(key)
    plt.imshow(masked_image)
    plt.show()
```
